```python
def durationFromBPM(bpm):
    print ("If a quarter equals", bpm, "then the duration of one quarter \n would be ", 60/bpm, " second(s).")
```

```python
durationFromBPM(60)
```

```python
durationFromBPM(120)
```

```python
durationFromBPM(180)
```

```python
def BPMfromDuration (duration):
    print ("the bmp would be ", 60/duration, " beats per minute.")
```

```python
BPMfromDuration(0.5)
```

```python
BPMfromDuration(2)
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

def tempos(quarter):
    quarter = quarter
    half = quarter / 2
    eighth= quarter * 2
    print ("quarter note = ", quarter, "beats per minute")
    print("quarter note duration = ", round(60 /quarter, 2), "seconds \n")
    print ("half note = ", half, "beats per minute")
    print("half note duration = ", round( 60/half, 2), "seconds\n")
    print ("eighth note = ", eighth, "beats per minute")
    print("eighth note duration = ", round( 60/eighth, 2), "seconds")

quarter = widgets.IntSlider(
    value=120,
    min=0,
    max=250,
    step=4,
    description='quarter:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)


widgets.interactive (tempos, quarter = quarter)
```
